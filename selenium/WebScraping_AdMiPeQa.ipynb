Then, import the required libraries.

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.common.exceptions import *
import pandas as pd
```

The website link and key word of search item were declared. 

```python
Lazada_url = 'https://www.lazada.com.my'
search_item = 'Milo Activ Go 1kg' 
```

Connect to the Chrome webdriver and search the url of Lazada.

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get(Lazada_url)
```

Find the product.

```python
p = driver.find_element(By.ID, 'q')
p.send_keys(search_item)
p.submit()
```

Get the product title, price, location and review by their classes. 

```python
item_titles = driver.find_elements(By.CLASS_NAME, 'RfADt')
item_prices = driver.find_elements(By.CLASS_NAME, 'ooOxS')
item_locations = driver.find_elements(By.CLASS_NAME, 'oa6ri')
item_reviews = driver.find_elements(By.CLASS_NAME, 'qzqFw')
```

Next, we unpack the item_titles, item_prices, item_locations and item_reviews variables onto lists:

```python
# Initialize empty lists
titles_list = []
prices_list = []
locations_list = []
reviews_list = []

# Loop over the item_titles and item_prices
for title in item_titles:
    titles_list.append(title.text)
for price in item_prices:
    prices_list.append(price.text)
for location in item_locations:
    locations_list.append(location.text)
for review in item_reviews:
    reviews_list.append(review.text)
```

We convert the four lists into a dataframe:

```python
df = pd.DataFrame(zip(titles_list, prices_list, locations_list, reviews_list), columns=['ItemName', 'Price', 'Location', 'Review'])
df
```

Clean the data.

```python
df['Price'] = df['Price'].str.replace('RM', '').astype(float)
df['Review'] = df['Review'].str.replace('(', '')
df['Review'] = df['Review'].str.replace(')', '')
df
```

Last, we export the data to a csv file

```python
# Export the dataframe to csv file
df.to_csv('Lazada_Milo.csv', index=False)
```
